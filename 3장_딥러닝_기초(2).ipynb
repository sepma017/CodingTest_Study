### 순전파를 코드로 구현
- 넘파이의 dot 함수를 이용해 다음과 같이 구현할 수 있다.

```python
# x: 입력 행렬 w: 가중치 행렬 b: 편향 벡터
u = np.dot(x, w) + b
# y: 출력 행렬 f: 활성화 함수
y = f(u)
```

- 편향 $b$는 벡터 => 이 벡터를 세로 방향으로 브로드캐스팅하면 행렬처럼 연산된다.
